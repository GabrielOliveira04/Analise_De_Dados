### Importação de Módulos e Arquivos

```python
import pandas as pd


funcionarios_df= pd.read_csv('CadastroFuncionarios.csv', sep=';',decimal=',')

clientes_df=pd.read_csv('CadastroClientes.csv',sep=';',decimal=',')

servicos_df=pd.read_excel('BaseServiçosPrestados.xlsx')

#Retirando as colunas Estado Civil e Cargo da tabela de funcionarios

funcionarios_df = funcionarios_df.drop(['Estado Civil', 'Cargo'], axis =1)

display(funcionarios_df)
display(clientes_df)
display(servicos_df)
```

### Folha Salarial

```python
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']

print('Total da Folha Salarial Mensal é de R${:,}'.format(funcionarios_df['Salario Total'].sum()))
```

### Faturamento da Empresa

```python
#Para o faturamento da empresa vamos pegar as duas tabelas a de clientes e a de serviço
#1° Vamos pegar o cliente 1 e multiplicar pelo valor contrato mensal que é 540 vezes 14 que é o tempo total de contrato
#Depois que calcular o faturamento de um cliente é so somar todos os faturamentos que chego no faturamento total
```

```python
faturamentos_df = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente','Valor Contrato Mensal']],on='ID Cliente')
display(faturamentos_df)

#Agora é só Multiplicar essas duas colunas tempo total de contrato e valor contrato mensal

faturamentos_df['Faturamento Total'] = faturamentos_df['Tempo Total de Contrato (Meses)'] * faturamentos_df['Valor Contrato Mensal']
print('Faturamento Total: R${:,}'.format(sum(faturamentos_df['Faturamento Total'])))
```

### % De Funcionarios  que Fecharam Contrato

```python
#1° Calcular quantidade  Total de funcionarios que tem, ai conta através da coluna ID Funcinario
#Para Quantos  funcionarios fecharam contrato é através da tabela de serviços do I Funcionario
#Porém tem funncionario que fechou mais de um contrato, dai vou tirar o funcionario duplicados
#Vou pegar quantos funcionarios fecharam contrato e dividir por funcionarios totais para saber o percentual
```

```python
funcionario_ctt = len(servicos_df['ID Funcionário'].unique())
funcionario_total = len(funcionarios_df['ID Funcionário'])
print('Percentual Funcionários  Fecharam Contrato: {:.2%}'.format(funcionario_ctt/funcionario_total))
```

### Qtde Contratos por Área

```python
#Dentro do merge botar o primeiro a tabela que vai receber a informações e dentro do parênteses botas as que vão das informações
contratos_area_df = servicos_df[['ID Funcionário']].merge(funcionarios_df[['ID Funcionário','Area']], on='ID Funcionário')
contratos_area_df  =contratos_area_df['Area'].value_counts()
print(contratos_area_df)
contratos_area_df.plot(kind='pie')
```

### Funcionários por Área

```python
funcionarios_area = funcionarios_df['Area'].value_counts()
print(funcionarios_area)
funcionarios_area.plot(kind='bar')
```

### Ticket Médio Mensal


```python
import matplotlib.pyplot as plt
ticket_medio = clientes_df['Valor Contrato Mensal'].mean()

print('Ticket Méio Mensal: R${:,.2f}'.format(ticket_medio))

# Plotando o valor médio como uma linha horizontal no gráfico
plt.axhline(y=ticket_medio, color='r', linestyle='--', label='Ticket Médio Mensal')
plt.legend()
plt.show()
```
